```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
global_electricity_df = pd.read_csv("https://raw.githubusercontent.com/owid/energy-data/master/owid-energy-data.csv")
global_electricity_df
```

```python
global_electricity_df['year'].max()
```

```python
recent_year = 2024
recent_electricity_df = global_electricity_df[global_electricity_df['year'] == recent_year]
recent_electricity_df
```

```python
df_renewables = recent_electricity_df[['country', 'year', 'iso_code', 'renewables_share_elec']].dropna()
df_renewables
```

```python
df_renewables_2 = df_renewables[df_renewables['iso_code'].str.len() == 3]
df_renewables_2
```

```python
top_10 = df_renewables_2.sort_values(by = 'renewables_share_elec', ascending = False).head(10)
top_10
```

```python
bottom_10 = df_renewables_2[df_renewables_2['renewables_share_elec'] > 0].sort_values(by = 'renewables_share_elec', ascending = True).head(10)
bottom_10
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top_10, x='renewables_share_elec', y='country', hue ='country', palette='Greens', legend=False)
plt.title('Top 10 Countries by Share of Renewables in Electricity Mix (2024)')
plt.xlabel('% of Electricity from Renewables')
plt.ylabel('Country')
plt.xlim(0, 100)
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.barplot(data=bottom_10, x='renewables_share_elec', y='country', hue = 'country', palette='Reds', legend=False)
plt.title('Bottom 10 Countries by Share of Renewables in Electricity Mix (2024)')
plt.xlabel('% of Electricity from Renewables')
plt.ylabel('Country')
plt.xlim(0, bottom_10['renewables_share_elec'].max()+ 0.5)
plt.show()
```
